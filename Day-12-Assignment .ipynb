```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your actual dataset)
data = {
    'Account_ID': [1, 2, 3, 4, 5],
    'Account_Type': ['Savings', 'Checking', 'Savings', 'Checking', 'Savings'],
    'Transaction_Amount': [2500, 1500, 3500, 200, 1000],
    'Branch': ['Uptown', 'Downtown', 'Uptown', 'Uptown', 'Downtown']
}

df = pd.DataFrame(data)

# Task 1: Bar plot of total sum of Transaction_Amount per Account_Type
transaction_sum_per_type = df.groupby('Account_Type')['Transaction_Amount'].sum()

# Plot the bar chart
plt.figure(figsize=(8, 6))
transaction_sum_per_type.plot(kind='bar', color='skyblue')
plt.title('Total Sum of Transaction_Amount per Account_Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=0)  # To make x-axis labels horizontal
plt.show()

# Task 2: Pie chart to show the percentage of transactions per Branch
transaction_count_per_branch = df['Branch'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
transaction_count_per_branch.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'gold', 'lightgreen'])
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')  # Hide the y-axis label
plt.show()
```
