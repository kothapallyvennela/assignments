```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\K.vennela\Downloads\Day_13_Pharma_data.csv")  # Ensure this file exists

# Data Cleaning
print("Initial data shape:", df.shape)
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
print("Data shape after cleaning:", df.shape)

# Bar plot: Total sales per region
plt.figure(figsize=(10, 5))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum, errorbar=None)  # Fix applied here
plt.title("Total Sales per Region")
plt.xticks(rotation=45)
plt.show()

# Scatter plot: Marketing Spend vs Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=df)
plt.title("Marketing Spend vs Sales")
plt.xlabel("Marketing Spend")
plt.ylabel("Sales")
plt.show()

# Boxplot: Drug Effectiveness across Age Groups
plt.figure(figsize=(8, 5))
sns.boxplot(x='Age_Group', y='Effectiveness', data=df)
plt.title("Drug Effectiveness by Age Group")
plt.show()

# Check if 'Product' column exists before plotting
if 'Product' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.lineplot(x='Trial_Period', y='Sales', hue='Product', data=df)
    plt.title("Sales Trend Over Trial Periods")
    plt.show()
else:
    print("Column 'Product' is missing from the dataset. Skipping line plot.")

# Heatmap: Correlation Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(df[['Sales', 'Marketing_Spend', 'Effectiveness']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Between Sales, Marketing Spend, and Effectiveness")
plt.show()

print("Analysis complete. Check visualizations for insights.")
```
